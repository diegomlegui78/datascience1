```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/diegomlegui78/datascience1/refs/heads/main/Employees_dataset1.csv"
df = pd.read_csv(url)
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
df_seleccion = df[['age', 'gender', 'KPIs_met_more_than_80', 'age', 'Salary']]
```

```python
df_seleccion.isna().sum()
```

```python
df_seleccion.dropna(inplace=True)
```

```python
df_encoded = pd.get_dummies(df_seleccion, columns=['gender'], drop_first=True)
```

En este caso vamos a predecir el personal que tiene un KPI mayor a 80. Que asignamos a la columna Y. Versus las variable gender, age y Salary

```python
X = df_encoded.drop(columns=['KPIs_met_more_than_80'])
y = df_encoded['KPIs_met_more_than_80']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
modelo_clas = DecisionTreeClassifier(random_state=42)
```

```python
modelo_clas.fit(X_train, y_train)
```

```python
y_pred = modelo_clas.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt=".0f")
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)
print(report)
```
