```python
import warnings
warnings.filterwarnings("ignore")

f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, figsize=(10,6), gridspec_kw={"height_ratios": (.15, .85)})
sns.boxplot(df['avg_training_score'], ax=ax_box, orient="h") # boxplot es diagrama de cajas
sns.histplot(df['avg_training_score'], ax=ax_hist, label='Puntuacion Promedio de Entrenamiento')
ax_box.set(yticks=[])
sns.despine(ax=ax_hist)
sns.despine(ax=ax_box, left=True)
ax_hist.legend()
```

```python
#rellenar los vacios
df['previous_year_rating'].fillna(df['previous_year_rating'].mode()[0], inplace = True)
df['education'].fillna(df['education'].mode()[0], inplace = True)
```

```python
# Comparar por departamento Edad/Cantidad
f, axs = plt.subplots(1, 2, figsize=(14, 6), gridspec_kw=dict(width_ratios=[4, 3]))
sns.scatterplot(data=df, y="department", x="age", hue="department", ax=axs[0],palette="ch:r=-.2,d=.3_r",sizes=(1, 8),legend=False,)
sns.histplot(data=df, y="department", hue="age", shrink=.8, alpha=.8, legend=False, ax=axs[1])
f.tight_layout()
```

```python
awards_by_channel_dept = df.groupby(['department', 'recruitment_channel'])['awards_won'].sum().reset_index()

# acumular en que canal hay mas premios
best_channel_by_dept = awards_by_channel_dept.loc[awards_by_channel_dept.groupby('department')['awards_won'].idxmax()]

print("Mejor Canal de reclutamiento/Departamento segun la cantidad de Premios Ganados:")
print(best_channel_by_dept)

plt.figure(figsize=(12, 6))
sns.barplot(x='department', y='awards_won', hue='recruitment_channel', data=awards_by_channel_dept)
plt.title('Premios Ganados por Departmento y Forma de Reclutamiento')
plt.xlabel('Departmento')
plt.ylabel('Total Premios Ganados')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#contar la cantidad de empleados con kpi mayor a 80 por educacion

# Filtrar empleados con KPI mayor a 80
df_kpi_80 = df[df['KPIs_met_more_than_80'] == 1]

# Agrupar por nivel de educaciÃ³n y contar la cantidad de empleados
education_kpi_count = df_kpi_80.groupby('education').count()
sns.histplot(data=education_kpi_count, y="education")

print("Cantidad de empleados con KPI mayor a 80 por nivel de educaciÃ³n:")
education_kpi_count
```
